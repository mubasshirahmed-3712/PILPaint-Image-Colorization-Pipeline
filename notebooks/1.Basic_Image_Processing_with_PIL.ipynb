<h2 style="text-align:center;">üß∞ Section 2 ‚Äì Load Image and Convert to NumPy Array</h2>

In this section, we‚Äôll load an image using the <code>PIL.Image.open()</code> method and convert it to a <code>NumPy</code> array using <code>np.asarray()</code>.

This allows us to directly inspect and manipulate pixel-level data.

- ‚úÖ Understand image shape and channel structure
- ‚úÖ Prepare image for transformations and filtering
- ‚úÖ Visualize using <code>matplotlib.pyplot</code>

<hr>


```python
# ‚úÖ Import required libraries
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# üì• Load image from input folder
img_path = '../input/3.jpg'
img = Image.open(img_path)

# ‚û§ Show original image
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
plt.show()
```

```python
# üîÑ Convert PIL image to NumPy array for pixel manipulation
car_arr = np.asarray(img)

# ‚û§ Print shape to understand dimensions (H x W x Channels)
print("‚úÖ NumPy array shape:", car_arr.shape)

# ‚û§ Optional: Preview the raw array (commented out if too large)
# print(car_arr)
```

<hr>

<h2 style="text-align:center;">üîç Section 3 ‚Äì Visualize RGB Channels Using NumPy</h2>

Each image is composed of 3 color channels:
- üî¥ Red
- üü¢ Green
- üîµ Blue

Using <code>NumPy</code>, we can extract these channels and visualize them individually. This is a great way to understand the structure of image data at a pixel level.

- ‚úÖ Slice 3D image array into separate channels
- ‚úÖ Use different colormaps for clarity
- ‚úÖ Display using <code>matplotlib</code>

<hr>


```python
# üîç Show RGB channels separately using array slicing
plt.figure(figsize=(16, 5))

# üî¥ Red Channel
plt.subplot(1, 3, 1)
plt.imshow(car_arr[:, :, 0], cmap='Reds')
plt.title("Red Channel")
plt.axis('off')

# üü¢ Green Channel
plt.subplot(1, 3, 2)
plt.imshow(car_arr[:, :, 1], cmap='Greens')
plt.title("Green Channel")
plt.axis('off')

# üîµ Blue Channel
plt.subplot(1, 3, 3)
plt.imshow(car_arr[:, :, 2], cmap='Blues')
plt.title("Blue Channel")
plt.axis('off')

plt.tight_layout()
plt.show()
```

<hr>

<h2 style="text-align:center;">üîÑ Section 4 ‚Äì Rotate and Flip the Image Using PIL</h2>

Image transformations such as rotation and flipping are often used for:
- üß™ Data augmentation in Machine Learning
- üé® Creative manipulations in photo editing
- üß≠ Changing the orientation for visual balance

We will apply the following:
- üîÅ 90¬∞ Rotation using <code>Image.rotate()</code>
- üîÑ Vertical Flip using <code>Image.transpose()</code>

<hr>


```python
# üîÅ Rotate the image by 90 degrees
rotated_img = img.rotate(90)
rotated_img.save("../output/rotate_90.jpg")

# üîÑ Flip the image vertically
flipped_img = img.transpose(Image.FLIP_LEFT_RIGHT)
flipped_img.save("../output/flip_top_bottom.jpg")
```

```python
# ‚û§ Display both transformed images
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(rotated_img)
plt.title("Rotated 90¬∞")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(flipped_img)
plt.title("Flipped Vertically")
plt.axis('off')

plt.tight_layout()
plt.show()
```

<hr>

<h2 style="text-align:center;">üé® Section 5 ‚Äì Apply Filters: Blur, Sharpen, Contour, and More</h2>

The <code>PIL.ImageFilter</code> module provides built-in filters to enhance or transform images. These are commonly used in:

- üé® Creative tools and editors
- üß™ Preprocessing for edge/feature detection
- üîç Improving visual clarity

We will apply the following filters:
- üîπ Blur
- üîπ Sharpen
- üîπ Detail
- üîπ Find Edges
- üîπ Contour

Each filter will be saved and displayed for comparison.

<hr>


```python
from PIL import ImageFilter

# üé® Dictionary of filters to apply
filters = {
    "blur": ImageFilter.BLUR,
    "sharpen": ImageFilter.SHARPEN,
    "detail": ImageFilter.DETAIL,
    "edges": ImageFilter.FIND_EDGES,
    "contour": ImageFilter.CONTOUR
}

# ‚û§ Apply, save, and store filtered images
filtered_images = {}

for name, filter_type in filters.items():
    filtered = img.filter(filter_type)
    filtered.save(f"../output/{name}.jpg")
    filtered_images[name] = filtered
```

```python
# üñºÔ∏è Display all filters in a grid
plt.figure(figsize=(15, 6))

for i, (name, image) in enumerate(filtered_images.items(), start=1):
    plt.subplot(2, 3, i)
    plt.imshow(image)
    plt.title(name.capitalize())
    plt.axis('off')

plt.tight_layout()
plt.show()
```
